```python
import os
from ultralytics import YOLO
import pandas as pd
import cv2

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Directory containing the trailers
trailer_dir = 'yt trailers'

# Function to process each trailer
def process_trailer(movie_id):
    # Get the trailer path
    trailer_path = os.path.join(trailer_dir, f"{movie_id}.mp4")
    
    # Create a dictionary to store the object frequencies
    object_counts = {}
    
    # Open the video file
    video = cv2.VideoCapture(trailer_path)
    
    # Get the total number of frames in the video
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Process every 5th frame
    for frame_idx in range(0, total_frames, 5):
        # Set the video position to the desired frame
        video.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        
        # Read the frame
        ret, frame = video.read()
        
        if ret:
            # Run YOLOv8 inference on the frame
            results = model(frame, verbose=False)  # Disable verbose output

            # Get the detected objects and their class IDs
            for result in results:
                boxes = result.boxes
                class_ids = boxes.cls.tolist()
                confidences = boxes.conf.tolist()
                
                # Count the frequency of each object class with confidence > 0.5
                for class_id, confidence in zip(class_ids, confidences):
                    if confidence > 0.5:
                        class_id = int(class_id)
                        object_counts[class_id] = object_counts.get(class_id, 0) + 1
    
    # Release the video file
    video.release()
    
    return object_counts

# Get the unique class IDs (assuming COCO dataset with 80 classes)
class_ids = list(range(80))

# Add new columns for each class ID and initialize with 0
for class_id in class_ids:
    sample_df[str(class_id)] = 0

# Process each row of the DataFrame
for index, row in sample_df.iterrows():
    movie_id = row['movieId']
    object_counts = process_trailer(movie_id)
    
    # Update the row with the object frequencies
    for class_id, count in object_counts.items():
        sample_df.at[index, str(class_id)] = count

# Print the updated DataFrame
print(sample_df.head())
```

```python
import os
from ultralytics import YOLO
import pandas as pd
import cv2
from concurrent.futures import ThreadPoolExecutor, as_completed

model = YOLO('yolov8n.pt')
trailer_dir = 'yt trailers'

def process_trailer(movie_id):
    print(movie_id)
    trailer_path = os.path.join(trailer_dir, f"{movie_id}.mp4")  
    object_counts = {}   
    video = cv2.VideoCapture(trailer_path)  
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))   
    for frame_idx in range(0, total_frames, 5):
        video.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)       
        ret, frame = video.read()       
        if ret:
            results = model(frame, verbose=False)         
            for result in results:
                boxes = result.boxes
                class_ids = boxes.cls.tolist()
                confidences = boxes.conf.tolist()
              
                for class_id, confidence in zip(class_ids, confidences):
                    if confidence > 0.5:
                        class_id = int(class_id)
                        object_counts[class_id] = object_counts.get(class_id, 0) + 1
    
    video.release()   
    return movie_id, object_counts

class_ids = list(range(80))

# Add new columns for object counts, initializing them to zero
for class_id in class_ids:
    sample_df[str(class_id)] = 0

# List of movie IDs
movie_ids = sample_df['movieId'].tolist()

# Use ThreadPoolExecutor to process trailers in parallel
with ThreadPoolExecutor(max_workers=30) as executor:
    future_to_movie_id = {executor.submit(process_trailer, movie_id): movie_id for movie_id in movie_ids}
    
    for future in as_completed(future_to_movie_id):
        movie_id, object_counts = future.result()
        index = sample_df[sample_df['movieId'] == movie_id].index[0]
        
        for class_id, count in object_counts.items():
            sample_df.at[index, str(class_id)] = count

print(sample_df)
```
