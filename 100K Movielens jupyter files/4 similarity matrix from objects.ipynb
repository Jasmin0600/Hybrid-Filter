```python
All_parts_objects = pd.read_csv(r'All_parts_objects.csv', index_col=2)
```

```python
columns_to_keep = [str(i) for i in range(80)]
movies_features = All_parts_objects[columns_to_keep]
movies_features
```

```python
ratings['rating'] = ratings['rating'].apply(lambda x: 1 if x > 3.5 else 0)
```

```python
import pandas as pd
from sklearn.preprocessing import normalize
```

```python
# Assuming 'ratings' is your DataFrame
user_movie_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating')
user_movie_matrix
```

```python
user_movie_matrix.info()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Fill NaN values with 0.5 in user_movie_matrix
user_movie_matrix = user_movie_matrix.fillna(0.5)

# Calculate the cosine similarity between movies based on their features
movie_similarity = cosine_similarity(movies_features)

# Convert the similarity matrix to a DataFrame
movie_similarity_df = pd.DataFrame(movie_similarity, index=movies_features.index, columns=movies_features.index)

# Function to predict ratings using KNN and item-item collaboration
def predict_ratings(user_ratings, k=10):
    user_id = user_ratings.name
    
    # Get the movies rated by the user
    rated_movies = user_ratings[user_ratings != 0.5].index
    
    # Calculate the weighted average of similar movies' ratings
    predicted_ratings = []
    for movie_id in user_movie_matrix.columns:
        if user_ratings[movie_id] == 0.5:
            similar_movies = movie_similarity_df[movie_id].sort_values(ascending=False)[1:k+1]
            similar_ratings = user_ratings[similar_movies.index]
            weights = similar_movies.values
            predicted_rating = np.dot(similar_ratings, weights) / weights.sum()
            predicted_ratings.append(predicted_rating)
        else:
            predicted_ratings.append(user_ratings[movie_id])
    
    return pd.Series(predicted_ratings, index=user_movie_matrix.columns, name=user_id)

# Predict ratings for all users
predicted_matrix = user_movie_matrix.apply(predict_ratings, axis=1)

# Convert the predicted ratings to binary (0 or 1) based on a threshold
threshold = 0.7  # Adjust the threshold based on your requirements
predicted_binary = (predicted_matrix >= threshold).astype(int)

# Evaluate the predictions using confusion matrix, precision, and recall
true_ratings = user_movie_matrix.values.flatten()
predicted_ratings = predicted_binary.values.flatten()

cm = confusion_matrix(true_ratings, predicted_ratings)
precision = precision_score(true_ratings, predicted_ratings)
recall = recall_score(true_ratings, predicted_ratings)

print("Confusion Matrix:")
print(cm)
print("Precision:", precision)
print("Recall:", recall)
```
