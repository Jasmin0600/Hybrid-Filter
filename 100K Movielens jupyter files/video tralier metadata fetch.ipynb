```python
import os
from moviepy.editor import VideoFileClip
import pandas as pd
import numpy as np

# Create empty lists to hold the metadata
durations = []
fps = []
widths = []
heights = []
aspect_ratios = []

# Iterate over the 'movieId' column of the dataframe
for movie_id in yt_only_tralier['movieId']:
    # Construct the file name using the movie_id
    file_name = f"{movie_id}.mp4"
    
    # Construct the path to the video file inside the 'yt trailers' folder
    file_path = os.path.join("yt trailers", file_name)
    
    # Initialize variables for metadata
    duration = None
    video_fps = None
    width = None
    height = None
    aspect_ratio = None

    try:
        # Load the video file to get its metadata
        with VideoFileClip(file_path) as video:
            duration = int(video.duration)
            video_fps = int(video.fps)
            width = int(video.size[0])
            height = int(video.size[1])
            aspect_ratio = round(width / height, 2)
    except (OSError, IndexError):
        print(f"Failed to extract metadata for video file {file_name}.")
    
    # Append the metadata to the lists
    durations.append(duration)
    fps.append(video_fps)
    widths.append(width)
    heights.append(height)
    aspect_ratios.append(aspect_ratio)

# Add the metadata to the dataframe as new columns
yt_only_tralier['Durations'] = durations
yt_only_tralier['FPS'] = fps
yt_only_tralier['Width'] = widths
yt_only_tralier['Height'] = heights
yt_only_tralier['Aspect Ratio'] = aspect_ratios

# Replace None values with NaN for consistency
yt_only_tralier.fillna(value=np.nan, inplace=True)
```
