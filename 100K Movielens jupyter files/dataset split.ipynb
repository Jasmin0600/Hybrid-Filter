```python
import numpy as np 
import pandas as pd 
```

```python
yt_only_trailer = pd.read_csv('CSV files/yt_only_trailer_2.csv')
```

```python
yt_only_trailer.info()
```

```python
# Sort the DataFrame by 'movieId' column
yt_only_trailer = yt_only_trailer.sort_values('movieId')

# Reset the index based on the sorted order
yt_only_trailer = yt_only_trailer.reset_index(drop=True)

# Display the updated DataFrame information
yt_only_trailer.info()
```

```python
yt_only_trailer
```

```python
# Calculate the number of rows for each part
total_rows = len(yt_only_trailer)
part1_rows = int(total_rows * 0.3)
part2_rows = int(total_rows * 0.2)
part3_rows = int(total_rows * 0.2)
part4_rows = int(total_rows * 0.15)
part5_rows = total_rows - part1_rows - part2_rows - part3_rows - part4_rows

# Adjust the number of rows in the last part to include any remaining rows
part5_rows += total_rows - (part1_rows + part2_rows + part3_rows + part4_rows + part5_rows)

# Divide the dataset into 5 parts
part1 = yt_only_trailer.iloc[:part1_rows]
part2 = yt_only_trailer.iloc[part1_rows:part1_rows+part2_rows]
part3 = yt_only_trailer.iloc[part1_rows+part2_rows:part1_rows+part2_rows+part3_rows]
part4 = yt_only_trailer.iloc[part1_rows+part2_rows+part3_rows:part1_rows+part2_rows+part3_rows+part4_rows]
part5 = yt_only_trailer.iloc[part1_rows+part2_rows+part3_rows+part4_rows:]

# Display the information of each part
print("Part 1:")
part1.info()
print("\nPart 2:")
part2.info()
print("\nPart 3:")
part3.info()
print("\nPart 4:")
part4.info()
print("\nPart 5:")
part5.info()

# Verify the total number of rows in all parts
total_rows_in_parts = len(part1) + len(part2) + len(part3) + len(part4) + len(part5)
print(f"\nTotal rows in all parts: {total_rows_in_parts}")
print(f"Total rows in the original dataset: {total_rows}")
```

```python
# Save each part as a CSV file
part1.to_csv('part1.csv', index=False)
part2.to_csv('part2.csv', index=False)
part3.to_csv('part3.csv', index=False)
part4.to_csv('part4.csv', index=False)
part5.to_csv('part5.csv', index=False)
```
